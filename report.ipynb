## 数据可视化对于论文的重要性

## 用 Python 怎么做数据可视化

第三方库为 Python 的数据分析处理、数据可视化提供支持。先分析和处理数据，完了再将结果呈现，所以数据的处理与可视化其实是有先后次序的，二者关系紧密。所以提到数据可视化就不得不先提数据分析处理。

用 Python 做数据处理，我们常用以下框架：

- Pandas：数据预处理、缺失值填充、分组统计等。
  https://pandas.pydata.org/

- NumPy：矩阵运算、科学计算基础。
  https://numpy.org/

- SciPy：假设检验、傅里叶变换、优化问题。
  https://scipy.org/

- StatsModels：线性回归、ARIMA 模型等。
  https://www.statsmodels.org/stable/index.html

用上述库处理完数据，再结合一些数据可视化框架​​来输出数据为图表：

- Matplotlib：快速绘制简单图表或需要精细控制的场景。
  https://matplotlib.org/

- Seaborn：探索性数据分析（EDA）、统计关系可视化。
  https://seaborn.pydata.org/

- Plotly / Plotly Express​：仪表盘、交互式报告。
  https://plotly.com/

- Bokeh：动态仪表盘、流数据可视化。
  https://docs.bokeh.org/en/latest/

- Altair：数据探索、自动化报告。
  https://altair-viz.github.io/

- Pygal：网页展示、打印输出。
  https://www.pygal.org/en/stable/

实际使用时，Pandas和NumPy是最常用的，SciPy偶尔用用；做可视化绘图时一般用Matplotlib绘制简单的样图，基本看看数据处理得是否正常、是否符合预期。如果要画更美观的图以用于正式论文，再考虑改用Plotly、Seaborn，其他的可视化框架就很少用了，不过学有余力的话也可以去了解一下，有些框架是支持动态交互的，比如Pygal；有的框架还能画更复杂的图表，甚至地图、交通线路图等复杂模型都能描绘，如Altair。

但具体到做数模竞赛的话，把基础框架学会用会的话就够了，比如做23年国赛的C题，最多用到Plotly框架，其后的不必过多了解。

## 基本绘图：Matplotlib基础用法

（结合Matplotlib official documentation来演示）

Matplotlib这个框架的设计初衷就是模拟MATLAB的绘图功能。所以很多方面和MATLAB非常像，如果有MATLAB经验的话能够很快上手。

```python
# 简易折线图

import matplotlib.pyplot as plt # 用这行语句引入 matplotlib 框架

plt.plot([1, 2, 3, 4])          # 描点，list里面的值描述x轴位置对应的y值
plt.ylabel('some numbers')      # 设置y轴标签
plt.show()                      # 显示折线图
```

```python
import matplotlib.pyplot as plt

plt.plot([1, 4, 7, 9], [1, 4, 9, 16])   # 另一种用法，同时指定x坐标和y坐标
plt.show()
```

```python
# 指定折线图的样式

import matplotlib.pyplot as plt

# 后面跟一个样式描述 'ro'
# 'r' 指 red 红色，'o' 表示只画圆点而不连线
plt.plot([1, 4, 7, 9], [1, 4, 9, 16], 'ro')
# 指定坐标轴范围，x轴0~6，y轴0~20
plt.axis((0, 10, 0, 20))
plt.show()
```

```python
# matplotlib 结合数据处理框架来使用，以numpy为例

import numpy as np
import matplotlib.pyplot as plt

# NumPy 采集样例数据
t = np.arange(0., 5., 0.2)  # 0~5 左开右闭区间，每隔0.2采集一个样本点

# 复杂化描述绘图
# 每描述一条线，传3个参数：x值、y值、样式描述
# t**2表示t的平方，t**3表示t的立方
# 样式：'--'表示虚线，'s'表示正方形形状的离散点，'^'表示三角形形状的离散点
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
# 一个更复杂的例子：绘制散点图

import numpy as np
from matplotlib import pyplot as plt

# 数据由NumPy随机采样生成
data = {
    'a': np.arange(50),
    'c': np.random.randint(0, 50, 50),
    'd': np.random.randn(50)
}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

# data数据集从a到b的散点图
# c：点的颜色，d：点的大小
plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
# 绘制多图并列

from matplotlib import pyplot as plt

names = ['a', 'b', 'c']
values = [1, 10, 100]

plt.figure(figsize=(9, 3))  # 画布整体大小：9英尺*3英尺

# 第一张图：柱状图
plt.subplot(1, 3, 1)        # 1行3列，当前图在第1张
plt.bar(names, values)
plt.title('Bar Chart')

# 第二张图：散点图
plt.subplot(1, 3, 2)        # 1行3列，当前图在第2张
plt.scatter(names, values)
plt.title('Scatter Plot')

# 第三张图：折线图
plt.subplot(1, 3, 3)        # 1行3列，当前图在第3张
plt.plot(names, values)
plt.title('Line Chart')

plt.show()
```

```python
# 绘制一些实验研究中的图像，如物理学中的阻尼振荡图像

import numpy as np
from matplotlib import pyplot as plt

def f(t):
    return np.exp(-t) * np.cos(2 ** np.pi * t)

t1 = np.arange(0., 5., 0.1)
t2 = np.arange(0., 5., 0.02)

plt.figure()

plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2 * np.pi * t2), 'r--')

plt.show()
```

```python
# 频率分布直方图的描绘

import numpy as np
from matplotlib import pyplot as plt

# 正态分布，均值100, 标准差15
mu, sigma = 100, 15
# 用NumPy生成随机数据
x = mu + sigma * np.random.randn(10000)

n, bins, patches = plt.hist(x, 50, density=True, facecolor='g', alpha=0.75)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Frequency Distribution Histogram')
plt.text(60, 0.025, r'$\mu={},\ \sigma={}$'.format(mu, sigma))
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

```python
# 对图像进行注释，写一些注释文字的例子

import numpy as np
from matplotlib import pyplot as plt

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2 * np.pi * t)

ax = plt.subplot(111)
# 绘制cos函数的图像
line, = plt.plot(t, s, lw=2)
# 注释这个函数的局部最大值点
plt.annotate(
    'local max', xy=(2, 1), xytext=(3, 1.5),
    arrowprops=dict(facecolor='black', shrink=0.05)
)
plt.ylim(-2, 2)
plt.show()
```

```python
# Matplotlib 还支持非线性坐标轴（坐标轴的值不是均匀分布的），下面是一些例子

import numpy as np
from matplotlib import pyplot as plt

np.random.seed(20250804)

# 用NumPy随机生成一些近似线性的数据
y = np.random.normal(loc=0.5, scale=0.4, size=100)
# 用NumPy的数据筛选功能，只选出0~1开区间内的数据
y = y[(y > 0) & (y < 1)]
# 对数据进行排序
y.sort()
x = np.arange(len(y))

plt.figure()

# 常规（线性）坐标轴
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)

# log 对数坐标轴
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)

# symlog (symmetric log) 对称对数坐标轴
plt.subplot(223)
plt.plot(x, y - y.mean())   # 中心线上移，能画出负半部分
plt.yscale('symlog', linthresh=0.01)
plt.title('symlog')
plt.grid(True)

# logit 坐标轴
plt.subplot(224)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)

# 稍微调整坐标轴大小，使其能够显示完全（主要是 logit 坐标轴的坐标范围有点大）
plt.subplots_adjust(
    top=0.92, bottom=0.08,
    left=0.10, right=0.95,
    hspace=0.25, wspace=0.35
)

plt.show()
```

## 进阶绘图框架：Seaborn简要演示

Matplotlib基本上可以实现绘制统计图、数据变化图线、函数图像、结合数据分析框架绘制图像等复杂情形。但有个缺点：画出的图像颜色和风格在主观上有点单调，直接粘贴进论文的话不够博人眼球、至少不容易让人眼前一亮。跟手上用铅笔打草稿的效果差不多，至少从整体上看论文，如果看到这种图像的话就像个草图。

怎么提升图像的和图表的美观度？想追求数据内容更加直观化，或者让数据对比方式更明显的话，进阶框架Seaborn是个不二之选。从下面的简要演示可以自己对比一下，Seaborn与Matplotlib所展现出的数据谁更直观化，能让读者一眼道出重点。

```python
# 导入Seaborn框架
import seaborn as sns

# 使用默认主题
sns.set_theme()

# 加载网络上的样例数据集
tips = sns.load_dataset("tips")

# 将数据集可视化
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size"
)
```

阅读Seaborn的API文档，可以发现Seaborn的底层其实隐含了Matplotlib的一些逻辑，也就是Seaborn是对Matplotlib的再次封装，暴露出更多概念抽象化的接口。这种抽象可以是统计分析者专注于思考对数据从不同角度、以不同方式的处理的问题上，借助Seaborn自身强大的功能，得到更佳的数据可视化效果。

同时得益于这种抽象能力，Seaborn可以为数据分析人员提供各个视角的数据可视化效果。比如，Seaborn可以帮助数据分析人员快速生成直方图、散点图、箱线图、热力图、等高线图、密度图、小提琴图、等等。这些图形的设计思路、数据处理逻辑都已经被Seaborn封装好，数据分析人员只需要简单调用相应的函数即可快速生成图形。举下面一个简单的例子，分别要求Seaborn以数据样本点集中区域为主体以及数据样本相关系数集中区域为主体画出趋势图，得到的效果如何。

```python
import seaborn as sns

dots = sns.load_dataset("dots")
sns.relplot(
    data=dots, kind="line",
    x="time", y="firing_rate", col="align",
    hue="choice", size="coherence", style="choice",
    facet_kws=dict(sharex=False)
)
```

Seaborn的功能除忠实描述数据性质外，还能在一定范围内预测数据的变化趋势，以及进行误差分析。

```python
# Seaborn做趋势估计

import seaborn as sns

fmri = sns.load_dataset("fmri")
sns.relplot(
    data=fmri, kind="line",
    x="timepoint", y="signal", col="region",
    hue="event", style="event"
)
```

```python
# Seaborn做误差分析

import seaborn as sns

tips = sns.load_dataset("tips")
sns.lmplot(data=tips, x="total_bill", y="tip", col="time", hue="smoker")
```

Seaborn还可对离散数据样本进行连续化的趋势估计，并基于此计算和绘制数据的经验累积分布函数（ECDF）。这已经是统计学中比较深的概念了，实际很少用到。

```python
import seaborn as sns

tips = sns.load_dataset("tips")
sns.displot(data=tips, x="total_bill", col="time", kde=True)
```

```python
import seaborn as sns

tips = sns.load_dataset("tips")
sns.displot(data=tips, kind="ecdf", x="total_bill", col="time", hue="smoker", rug=True)
```

Seaborn还可以对数据进行归类分析，展现不同类型（Category）数据的分布情况。也可使用不同的图形进行展示。

```python
import seaborn as sns

tips = sns.load_dataset("tips")
```

```python
# swarm图
sns.catplot(data=tips, kind="swarm", x="day", y="total_bill", hue="smoker")
```

```python
# violin图
sns.catplot(data=tips, kind="violin", x="day", y="total_bill", hue="smoker", split=True)
```

```python
# 条形图
sns.catplot(data=tips, kind="bar", x="day", y="total_bill", hue="smoker")
```

对于具有多种类型的数据，Seaborn可以根据其在不同归类方式下的不同类型绘制二维归类散点图。

```python
import seaborn as sns

penguins = sns.load_dataset("penguins")
sns.jointplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", hue="species")
```

对于更多归类方式，Seaborn提供了一种 pair plot 的方式来展现各对归类方式下的二维归类散点图。

```python
import seaborn as sns

penguins = sns.load_dataset("penguins")
sns.pairplot(data=penguins, hue="species")
```

除了提供抽象接口外，Seaborn也提供了一些原理层接口，以便数据分析者有需求时使用这些接口绘制更精密的图像。还是以上面的 pair grid 为例，我们可以借助原理层接口在散点图的基础之上绘制等高线。

```python
import seaborn as sns

penguins = sns.load_dataset("penguins")
g = sns.PairGrid(penguins, hue="species", corner=True)
g.map_lower(sns.kdeplot, hue=None, levels=5, color=".2")
g.map_lower(sns.scatterplot, marker="+")
g.map_diag(sns.histplot, element="step", linewidth=0, kde=True)
g.add_legend(frameon=True)
g.legend.set_bbox_to_anchor((.61, .6))
```

通过Seaborn的原理层接口，我们可以在数据可视化的既有基础之上进行更细粒度的控制，如描绘数据点的颜色梯度。

```python
import seaborn as sns

penguins = sns.load_dataset("penguins")
sns.set_theme(style="ticks", font_scale=1.25)
g = sns.relplot(
    data=penguins,
    x="bill_length_mm", y="bill_depth_mm", hue="body_mass_g",
    palette="crest", marker="x", s=100,
)
g.set_axis_labels("Bill length (mm)", "Bill depth (mm)", labelpad=10)
g.legend.set_title("Body mass (g)")
g.figure.set_size_inches(6.5, 4.5)
g.ax.margins(.15)
g.despine(trim=True)
```

Seaborn的底层是完全基于Matplotlib的，所以很多Matplotlib的功能都能在Seaborn上得到呈现，二者也有很多联动。如果熟练掌握Matplotlib的话，上手Seaborn应该会非常容易。通过这两个框架结合，可以快速探索数据并创建可以定制成出版物质量最终产品的图形，以及能够运用于特定论文上的数据的可视化描述。把Seaborn用熟了可以达到甚至超越Matlab作出来的图像（印象中Matlab的作图格式是固定的，不能自定义），以更多细节、更广维度的数据图表为论文内容引入更多亮点。

除了Matplotlib和Seaborn，Python还有很多数据可视化框架，甚至有[可以交互式呈现数据成分的框架](https://www.pygal.org/en/stable/)，具体见上面的总结。碍于篇幅这里就不展开讲了。Python社区有越来越多的新兴框架正在涌现，上述链接中的Pygal就是其中之一，正是得益于Python社区的开放包容，相较于Matlab等传统数据统计分析软件，功能更加强大、方便。
