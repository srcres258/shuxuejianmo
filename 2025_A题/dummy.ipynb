```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("附件/电站1发电数据_估计.csv", index_col=0)
df.head(20)
```

```python
for i, row in enumerate(df.itertuples()):
    kwh = row[1]
    if kwh == 0.0 and not row[0].endswith('00:00:00') and i > 0:
        last_kwh = df.iloc[i - 1, 0]
        if last_kwh > 0.0:
            df.iloc[i, 0] = last_kwh
```
