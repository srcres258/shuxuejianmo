```python
import pandas as pd
```

```python
df1 = pd.read_csv('附件/电站1发电数据_估计.csv', parse_dates=['时间'])
df2 = pd.read_csv('附件/电站1环境检测仪数据_估计.csv', parse_dates=['时间'])
df3 = pd.read_csv('附件/电站1天气数据_估计.csv', parse_dates=['时间'])
```

```python
merged_df = df1
for df in [df2, df3]:
    merged_df = pd.merge(merged_df, df, on='时间', how='outer')
```

```python
merged_df = merged_df.drop_duplicates()
merged_df = merged_df.dropna()
merged_df.head(20)
```
